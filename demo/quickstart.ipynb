```python
# Imports and fixes for the module not found error
import sys
import os

sys.path.append(os.path.abspath(".."))
import numpy as np

from numcompute.io import load_csv
from numcompute.preprocessing import Imputer, StandardScaler, OneHotEncoder
from numcompute.sort_search import topk, binary_search
from numcompute.rank import rank, percentile
from numcompute.stats import mean, std, histogram, quantile
from numcompute.metrics import accuracy, f1_score, mse
from numcompute.optim import grad
from numcompute.pipeline import Pipeline
from numcompute.benchmarking import benchmark

# Loading the data
X = load_csv("../test.csv", skip_header=1)
print("Raw Data:")
print(X)
print("========================================")
# Preprocessing Pipeline
pipe = Pipeline(
    [("impute", Imputer()), ("encode", OneHotEncoder()), ("scale", StandardScaler())]
)

X_processed = pipe.fit_transform(X)

print("Processed Data:")
print(X_processed)
print("========================================")

# Using the topk and binary search functions
arr = np.array([10, 5, 20, 15])

vals, idx = topk(arr, 2)
print("Top 2 values:", vals)

i, found = binary_search(np.sort(arr), 15)
print("Binary search (15):", i, found)
print("========================================")

# Using the rank and percentile functions
data = np.array([10, 20, 20, 30])

print("Ranks (average):", rank(data, method="average"))
print("50th percentile:", percentile(data, 50))
print("========================================")

# Using the statistics functions
arr = np.array([1, 2, 3, np.nan])

print("Mean:", mean(arr))
print("Std:", std(arr))

counts, edges = histogram(np.array([1, 2, 2, 3, 4]), bins=3)

print("Histogram counts:", counts)
print("Quantile (0.5):", quantile(np.array([10, 20, 30, 40]), 0.5))
print("========================================")

# Using the metrics functions
y_true = np.array([1, 0, 1, 1])
y_pred = np.array([1, 0, 0, 1])

print("Accuracy:", accuracy(y_true, y_pred))
print("F1 Score:", f1_score(y_true, y_pred))
print("MSE:", mse([1, 2], [1, 3]))
print("========================================")


# Usint the gradient function
def f(x):
    return x[0] ** 2 + x[1] ** 2


g = grad(f, np.array([2.0, 3.0]))

print("Gradient:", g)
print("========================================")


# Using the benchmark function
def loop_sum(x):
    total = 0
    for i in x:
        total += i
    return total


def vectorized_sum(x):
    return np.sum(x)


x = np.random.rand(1000000)

benchmark("Sum", vectorized_sum, loop_sum, x)

print()
print("Entire module executed successfully!")
```
